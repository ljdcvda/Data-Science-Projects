# Classifying digits using a neural network
## Reading the dataset

```python
from sklearn.datasets import load_digits
import pandas as pd
digits = load_digits()
```

```python
digits.keys()
```

```python
data = pd.DataFrame(digits.data)
labels= pd.Series(digits.target)
data
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig, axes = plt.subplots(2, 4)
count = 0
for row in range(2):
    for col in range(4):
        axes[row,col].imshow(data.iloc[count].values.reshape(8,8), cmap='gray_r')
        count+=1
```

## Performance of a k-nearest neighbors algorithm

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import KFold

def train(train_features, train_labels, k=5):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(train_features, train_labels)
    return knn

def test(model, test_features, test_labels):
    predictions = model.predict(test_features)
    accuracy = sum(predictions == test_labels)/len(predictions)
    return accuracy

def cross_validate(n=4, k=5):
    accuracies = []
    kf = KFold(n_splits = n)
    for train_index, test_index in kf.split(data):
        train_features, test_features = data.loc[train_index], data.loc[test_index]
        train_labels, test_labels = labels.loc[train_index], labels.loc[test_index]
        model = train(train_features, train_labels, k=k)
        overall_accuracy = test(model, test_features, test_labels)
        accuracies.append(overall_accuracy)
    return sum(accuracies)/len(accuracies)

cross_validate()
```

```python
k_list = list(range(1,11))
accuracies = []
for k in k_list:
    accuracies.append(cross_validate(k=k))

plt.plot(k_list, accuracies)
plt.xlabel('k')
```

This algorithm performs surprisingly well!

## Using a single hidden layer neural network
Let's explore how the number of nodes affects the accuracy on the test and training sets, this should indicate if/where overfitting happens

```python
from sklearn.neural_network import MLPClassifier
layer_sizes = [(8, ), (16, ), (32, ), (64, ), (128, ), (256, )]

def train_nn(train_features, train_labels, size):
    mlp = MLPClassifier(hidden_layer_sizes = size, activation='logistic')
    mlp.fit(train_features, train_labels)
    return mlp

train_accuracies = []
test_accuracies = []
for size in layer_sizes:
    kf = KFold(n_splits = 4)
    train_accuracies_folds = []
    test_accuracies_folds = []
    for train_index, test_index in kf.split(data):
        train_features, test_features = data.loc[train_index], data.loc[test_index]
        train_labels, test_labels = labels.loc[train_index], labels.loc[test_index]
        model = train_nn(train_features, train_labels, size)
        train_accuracies_folds.append(test(model, train_features, train_labels))
        test_accuracies_folds.append(test(model, test_features, test_labels))
    train_accuracies.append(sum(train_accuracies_folds)/len(train_accuracies_folds))
    test_accuracies.append(sum(test_accuracies_folds)/len(test_accuracies_folds))
```

```python
plt.plot([x[0] for x in layer_sizes], train_accuracies, 'r', label='train acc')
plt.plot([x[0] for x in layer_sizes], test_accuracies, 'g', label='test acc')
plt.legend()
```
