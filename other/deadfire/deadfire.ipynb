```python
import pandas as pd
url = "https://pillarsofeternity.gamepedia.com/List_of_unique_weapons_(Deadfire)"
# table_list = pd.read_html(url, attrs = {'class': 'va-table va-table-full va-table-center sortable'})
table_list = pd.read_html(url)
for table in table_list:
    print(table.iloc[0,1], table.shape)
```

```python
copied_string = "Arbalests Arquebuses Battle_axes Blunderbusses Clubs Crossbows Daggers Estocs Flails Great_swords Hatchets Hunting_bows Maces Morning_stars Pikes Pistols Pollaxes Quarterstaffs Rapiers Rods Sabres Scepters Spears Stilettos Swords Wands War_bows War_hammers"
weapon_types = copied_string.split()
weapon_types
```

```python
print(len(weapon_types), len(table_list))
```

```python
for weapon, table in zip(weapon_types, table_list):
    table['type'] = weapon
combined_table = pd.concat(table_list)
combined_table = combined_table.drop('Icon', axis=1)
combined_table['Name'] = combined_table['Name'].str.replace('\(Unique\)', '')
combined_table = combined_table.set_index('Name')
combined_table
```

```python
combined_table = combined_table.fillna('')
combined_table[combined_table['Enchantments'].str.contains('Spirit')].iloc[1]['Enchantments']
```

## Adding enchantments
The data in these tables does not contain possible additional enchantments which are described on individual weapon pages.
How to get that data? From those pages
How to get the pages url? They are contained within the original tables, as well as other wiki pages.
The pandas url table reader however ignores href by default
* We can scrape the page ourselves and store the links
* Or we can scrape a different page and combined with our table data
* Or we can read through our data nad try to compose the urls since they follow a rule...
Lets try that

```python
import re
for weapon in combined_table.index:
    print(weapon,'=====>', re.search(r"^[^(\[]*", weapon)[0])
```

Now let's see what happens when we fetch

```python
base_url = "https://pillarsofeternity.gamepedia.com/"
for weapon in combined_table.index:
    if 'Soulbound' not in weapon:
        try:
            url = base_url+ re.search(r"^[^(\[]*", weapon)[0]
            url = url.strip().replace(' ', '_')
            print(url)
            table_list = pd.read_html(url, encoding="utf8")
            list_of_enchantments = table_list[-1]['Effect'].unique()
            combined_table.loc[weapon,'Additional Enchantments'] = ' '.join(list_of_enchantments)
        except:
            print('er')
combined_table
```
